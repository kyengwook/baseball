```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# CSV 파일 불러오기
merged_df = pd.read_csv('/Users/kyengwook/Documents/Baseball/PB/홈런여부판단/merged_Smith,_Will_2024.csv')

# 각 팀별로 내 예측과 MLB 예측의 Accuracy, Precision, Recall, F1-score 계산

def calculate_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return accuracy, precision, recall, f1

# 각 팀별로 예측 값 비교
team_list = ['BAL','BOS','NYY','TB','TOR','CWS','CLE','DET','KC','MIN','HOU','LAA','OAK','SEA','TEX','ATL','MIA','NYM','PHI','WSH','CHC','CIN','MIL','PIT','STL','ARI', 'COL', 'LAD', 'SD', 'SF']
metrics = {}

for team in team_list:
    y_true = merged_df[f'{team.lower()}']  # MLB 예측 (소문자)
    y_pred = merged_df[f'예측_홈런_{team}']  # 내 예측
    accuracy, precision, recall, f1 = calculate_metrics(y_true, y_pred)
    
    metrics[team] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-score': f1
    }

# 결과 출력
metrics_df = pd.DataFrame(metrics).T
print(metrics_df)

# 결과를 CSV로 저장
output_metrics_file = '/Users/kyengwook/Documents/Baseball/PB/홈런여부판단/homerun_predictions_metrics.csv'
metrics_df.to_csv(output_metrics_file)

#print(f"모든 팀에 대한 예측 성능 지표가 저장된 파일이 생성되었습니다: {output_metrics_file}")
```

```python
##SVM모델##

import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# 구장 목록
team_list = ["LAD"]  # 예시로 여러 B구장 추가

# 학습된 모델 저장 딕셔너리
team_models = {}

# 각 구장에서의 모델 학습
for team in team_list:
    df1 = df[df['home_team'] == team]
    df1 = df1[df1['description'] == 'hit_into_play']
    
    # 홈런 여부 열 생성
    df1['홈런여부'] = (df1['events'] == 'home_run').astype(int)

    # 학습 데이터 입력(X)와 출력(Y) 정의
    x_train = df1[['plate_x', 'plate_z', 'hc_x', 'hc_y', 'hit_distance_sc', 'launch_speed', 'launch_angle', 'launch_speed_angle']]
    y_train = df1['홈런여부']

    # 결측치 처리
    columns_to_fill = ['plate_x', 'plate_z', 'hc_x', 'hc_y', 'hit_distance_sc', 'launch_speed', 'launch_angle', 'launch_speed_angle']
    for col in columns_to_fill:
        x_train[col] = x_train[col].fillna(x_train[col].mean())

    # SVM은 스케일링에 민감하므로 StandardScaler 사용
    scaler = StandardScaler()

    # SVM 모델 생성 및 하이퍼파라미터 그리드
    svm_model = SVC(probability=True, random_state=42)
    param_grid = {
        'svc__C': [0.1, 1, 10, 100],  # 규제 파라미터
        'svc__gamma': [0.001, 0.01, 0.1, 1],  # RBF 커널의 gamma 값
        'svc__kernel': ['rbf']  # 커널 방식
    }

    # 파이프라인 설정
    pipeline = Pipeline([
        ('scaler', scaler),  # 데이터 스케일링
        ('svc', svm_model)   # SVM 모델
    ])

    # GridSearchCV로 하이퍼파라미터 튜닝
    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        scoring='accuracy',
        cv=3,
        n_jobs=-1,
        verbose=0
    )

    # 모델 학습
    grid_search.fit(x_train, y_train)

    # 최적 모델 저장
    team_models[team] = grid_search.best_estimator_

    print(f"{team} 구장에서의 SVM 모델 학습 완료.")

# 특정 선수의 홈런 데이터 로드
# 예: 특정 선수의 시즌 데이터
#player_data = df[df['batter'] == '특정 선수 이름']  # 선수 이름 필터링
#player_home_runs = player_data[player_data['events'] == 'home_run']  # 홈런 데이터만 추출

# 입력 피처만 추출
player_home_runs_features = Ans[['plate_x', 'plate_z', 'hc_x', 'hc_y', 'hit_distance_sc', 'launch_speed', 'launch_angle', 'launch_speed_angle']]

# 결측치 처리
for col in player_home_runs_features.columns:
    player_home_runs_features[col] = player_home_runs_features[col].fillna(player_home_runs_features[col].mean())

# 결과 저장용 DataFrame
results = []

# 각 B구장에서 선수의 모든 홈런 예측
for b_team in team_list:
    if b_team in team_models:
        b_model = team_models[b_team]
        predictions = b_model.predict(player_home_runs_features)  # 모든 홈런에 대해 예측

        # 홈런 여부 결과 저장
        for idx, pred in enumerate(predictions):
            results.append({
                'B_구장': b_team,
                '타구_index': idx,
                '실제_홈런': 1,  # 선수의 타구는 모두 실제 홈런
                '예측_홈런': pred
            })

        print(f"{b_team} 구장에서 선수의 홈런 데이터를 예측했습니다.")
    else:
        print(f"{b_team} 구장에 대한 모델이 존재하지 않습니다.")

# 결과 DataFrame 생성
results_df = pd.DataFrame(results)

# 예측 결과 요약 출력
summary = results_df.groupby('B_구장').agg(
    총_타구수=('타구_index', 'count'),
    실제_홈런수=('실제_홈런', 'sum'),
    예측_홈런수=('예측_홈런', 'sum')
)
print("\n예측 결과 요약:")
print(summary)

# 결과 CSV로 저장
#results_df.to_csv('player_home_run_predictions.csv', index=False)
#summary.to_csv('home_run_summary.csv', index=True)
```
